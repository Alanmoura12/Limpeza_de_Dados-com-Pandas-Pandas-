```python
df = df.dropna()
```

#### consultar as categorias das colunas

```python
df['experience_level'].value_counts()
```

#### Alterando nome das colunas

```python
colunas = {
    'work_year' : 'ano',
     'experience_level' : 'nivel'   
   
}

df.rename(columns = colunas, inplace=True)

df.head()
```

#### Alterando nome dos Conteudos da colunas

```python
nome_conteudo = {
    'SE' : 'senior',
    'MI' : 'pleno',
    'EN' : 'junior',
    'EX' : 'excecutivo'
}

df['nivel'] = df['nivel'].replace(nome_conteudo)

df['nivel'].value_counts()
```

```python
df.info()
```

```python
df = df.assign(ano = df['ano'].astype('int64'))
```
